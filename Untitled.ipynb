```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.xlabel('Input Feature 1')
    plt.ylabel('Input Feature 2')
    plt.title('Perceptron Decision Boundary')
    plt.show()

# Define the input data for OR, AND, and XOR
X_or = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_or = np.array([0, 1, 1, 1])

X_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])

X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])

# Train and visualize the Perceptron for OR
print("OR Gate")
perceptron_or = Perceptron(max_iter=1000, eta0=0.1, random_state=0)
perceptron_or.fit(X_or, y_or)
plot_decision_boundary(X_or, y_or, perceptron_or)

# Train and visualize the Perceptron for AND
print("AND Gate")
perceptron_and = Perceptron(max_iter=1000, eta0=0.1, random_state=0)
perceptron_and.fit(X_and, y_and)
plot_decision_boundary(X_and, y_and, perceptron_and)

# Train and visualize the Perceptron for XOR (will not work)
print("XOR Gate")
perceptron_xor = Perceptron(max_iter=1000, eta0=0.1, random_state=0)
perceptron_xor.fit(X_xor, y_xor)
plot_decision_boundary(X_xor, y_xor, perceptron_xor)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.xlabel('Input Feature 1')
    plt.ylabel('Input Feature 2')
    plt.title('MLP Decision Boundary for XOR (Improved)')
    plt.show()

# Define input data for XOR
X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])

# Create an improved MLP model
mlp = MLPClassifier(hidden_layer_sizes=(5, 5),  # Two hidden layers with 5 neurons each
                    activation='relu',         # Use ReLU activation function
                    solver='adam',             # Use Adam optimizer
                    max_iter=2000,             # Increase the number of iterations
                    learning_rate_init=0.01,   # Set a learning rate
                    random_state=0)

# Fit the model
mlp.fit(X_xor, y_xor)

# Predicting and checking accuracy
predictions = mlp.predict(X_xor)
print("Predictions:", predictions)
print("Accuracy:", mlp.score(X_xor, y_xor))

# Plot the decision boundary
plot_decision_boundary(X_xor, y_xor, mlp)
```
